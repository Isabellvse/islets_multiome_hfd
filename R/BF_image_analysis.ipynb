```python
data_raw = Path(here("data-raw/images"))
paths = list(data_raw.glob("*"))

def base_key(p: Path):
    return p.name.replace(" v2", "").replace(" v2.zip", "").replace(".zip", "")

# 1. group by logical identity
groups = defaultdict(list)
for p in paths:
    groups[base_key(p)].append(p)

final = []

for ps in groups.values():
    # 2. prefer v2
    v2 = [p for p in ps if "v2" in p.name]
    candidates = v2 if v2 else ps

    # 3. if duplicates remain, prefer non-zip
    non_zip = [p for p in candidates if p.suffix != ".zip"]
    candidates = non_zip if non_zip else candidates

    # 4. pick deterministic single result
    final.append(sorted(candidates)[0])
```

```python
clean_nd2_by_folder = {}

for folder in final:
    folder_name = folder.name
    remove_set = set(files_to_remove.get(folder_name, []))

    kept_files = [
        p.name
        for p in folder.glob("*.nd2")
        if p.name not in remove_set
    ]

    clean_nd2_by_folder[folder_name] = kept_files
```

```python
import subprocess
import sys
import psutil
import os
from pathlib import Path
from pyhere import here

# Open log file
log_file = open(str(here('log_insulin_masks.txt')), 'w')

# Process all folders
for folder_name, image_list in clean_nd2_by_folder.items():
    msg = f"\n{'='*50}\nExtracting insulin masks: {folder_name}\n{'='*50}"
    print(msg)
    log_file.write(msg + '\n')
    
    # Process 40 images at a time
    for i in range(0, len(image_list), 25):
        batch = image_list[i:i+25]
        processes = []
        
        # Start all processes (non-blocking)
        for image_name in batch:
            p = subprocess.Popen([
                sys.executable, '-c',
                f"""
import sys
sys.path.insert(0, '{os.getcwd()}')
import functions as f
from pathlib import Path
from pyhere import here

num_masks = f.extract_insulin_masks(
    path=str(here("data_raw/images")),
    folder_name='{folder_name}',
    image_name='{Path(image_name).stem}'
)
print(f"Extracted {{num_masks}} masks")
"""
            ])
            processes.append((image_name, p))
        
        # Wait for all to finish
        for image_name, p in processes:
            p.wait()
            msg = f"  ✓ {Path(image_name).stem} completed"
            print(msg)
            log_file.write(msg + '\n')
        
        # Check memory after batch
        mem = psutil.Process(os.getpid()).memory_info().rss / 1024 / 1024
        msg = f"  Batch {i//40}: {mem:.1f} MB"
        print(msg)
        log_file.write(msg + '\n')
        log_file.flush()
    
    msg = f"✓ {folder_name} finished\n"
    print(msg)
    log_file.write(msg + '\n')

log_file.close()
print("Done! Check log_insulin_masks.txt")
```
